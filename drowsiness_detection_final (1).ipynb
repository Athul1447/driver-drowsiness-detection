```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from PIL import Image
import pygame

pygame.mixer.init()

model = load_model('model.keras')


cap = cv2.VideoCapture(0)


closed_eyes_counter = 0  
yawning_counter = 0      
alarm_played = False     # avoid multiple alarm plays

# Function to play alarm sound
def play_alarm():
    try:
        pygame.mixer.music.load(r"C:\Users\ATHUL\Downloads\eyes.mp3") 
        pygame.mixer.music.set_volume(1.0)  
        pygame.mixer.music.play()
    except pygame.error as e:
        print(f"Error playing sound: {e}")

# Main loop
while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Could not capture frame from webcam.")
        break
    frame_resized = cv2.resize(frame, (420, 340))
    # Convert the frame to RGB and preprocess
    img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    img = Image.fromarray(img)
    img = img.resize((80,80))  # Resize to 80x80 pixels
    img = np.array(img) / 255.0  # Normalize the image

    # Make a prediction with the model
    result = model.predict(img[np.newaxis, ...])  # Add batch dimension
    predicted_label_index = np.argmax(result)

    # Update counters based on predictions
    if predicted_label_index == 0:  # Closed Eyes
        closed_eyes_counter += 1
        yawning_counter = 0
    elif predicted_label_index == 3:  # Yawning Detected
        yawning_counter += 1
        closed_eyes_counter = 0
    else:  # Open Eyes or No Yawn
        closed_eyes_counter = 0
        yawning_counter = 0
        alarm_played = False  # Reset alarm if no drowsiness is detected

    # Trigger alarm if thresholds are exceeded
    if closed_eyes_counter > 30 or yawning_counter > 30:  # ~1 second (assuming ~30 FPS)
        if not alarm_played:
            play_alarm()
            alarm_played = True

    # Display predictions on the frame
    if closed_eyes_counter > 0:  # Prioritize "Closed Eyes"
        cv2.putText(frame, 'Eyes Closed', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    elif yawning_counter > 0:  # Check for yawning only if eyes are not closed
        cv2.putText(frame, 'Yawning Detected', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)
    else:  # Normal state (e.g., open eyes)
        if predicted_label_index == 1:
            cv2.putText(frame, 'No Yawn Detected', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        elif predicted_label_index == 2:
            cv2.putText(frame, 'Open Eyes', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

    # Indicate drowsiness if thresholds are exceeded
    if closed_eyes_counter > 30 or yawning_counter > 30:
        cv2.putText(frame, 'Drowsiness Detected', (10, 150), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    # Display the frame
    cv2.imshow("Drowsiness Detection", frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
pygame.mixer.music.stop()
pygame.quit()
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from PIL import Image
import pygame
import threading
import tkinter as tk
from tkinter import messagebox

import os
import sys
import pygame

# Suppress Pygame startup message
with open(os.devnull, "w") as fnull:
    old_stdout = sys.stdout
    sys.stdout = fnull
    try:
        pygame.mixer.init()
    finally:
        sys.stdout = old_stdout


# Load the model
model = load_model('model.keras')

def play_alarm():
    """Function to play alarm sound."""
    try:
        pygame.mixer.music.load(r"C:\\Users\\ATHUL\\Downloads\\eyes.mp3")
        pygame.mixer.music.set_volume(1.0)
        pygame.mixer.music.play()
    except pygame.error as e:
        print(f"Error playing sound: {e}")

def start_detection():
    """Function to start drowsiness detection."""
    global detection_running, cap
    detection_running = True

    closed_eyes_counter = 0
    yawning_counter = 0
    alarm_played = False

    # Initialize video capture
    cap = cv2.VideoCapture(0)
    def detection_loop():
        nonlocal closed_eyes_counter, yawning_counter, alarm_played
        while detection_running:
            ret, frame = cap.read()
            if not ret:
                print("Error: Could not capture frame from webcam.")
                break

            frame_resized = cv2.resize(frame, (420, 340))
            # Convert the frame to RGB and preprocess
            img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            img = Image.fromarray(img)
            img = img.resize((80, 80))  # Resize to 80x80 pixels
            img = np.array(img) / 255.0  # Normalize the image

            # Make a prediction with the model
            result = model.predict(img[np.newaxis, ...])  # Add batch dimension
            predicted_label_index = np.argmax(result)

            # Update counters based on predictions
            if predicted_label_index == 0:  # Closed Eyes
                closed_eyes_counter += 1
                yawning_counter = 0
            elif predicted_label_index == 3:  # Yawning Detected
                yawning_counter += 1
                closed_eyes_counter = 0
            else:  # Open Eyes or No Yawn
                closed_eyes_counter = 0
                yawning_counter = 0
                alarm_played = False

            # Trigger alarm if thresholds are exceeded
            if closed_eyes_counter > 30 or yawning_counter > 30:  # ~1 second (assuming ~30 FPS)
                if not alarm_played:
                    play_alarm()
                    alarm_played = True

            # Display predictions on the frame
            if closed_eyes_counter > 0:  # Prioritize "Closed Eyes"
                cv2.putText(frame, 'Eyes Closed', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
            elif yawning_counter > 0:  # Check for yawning only if eyes are not closed
                cv2.putText(frame, 'Yawning Detected', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)
            else:  # Normal state (e.g., open eyes)
                if predicted_label_index == 1:
                    cv2.putText(frame, 'No Yawn Detected', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
                elif predicted_label_index == 2:
                    cv2.putText(frame, 'Open Eyes', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

            # Indicate drowsiness if thresholds are exceeded
            if closed_eyes_counter > 30 or yawning_counter > 30:
                cv2.putText(frame, 'Drowsiness Detected', (10, 150), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

            # Display the frame
            cv2.imshow("Drowsiness Detection", frame)

            # Exit if 'q' is pressed
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        stop_detection()

    threading.Thread(target=detection_loop, daemon=True).start()

def stop_detection():
    """Function to stop drowsiness detection."""
    global detection_running, cap
    detection_running = False

    if cap:
        cap.release()
    cv2.destroyAllWindows()
    pygame.mixer.music.stop()

# Create the Tkinter window
window = tk.Tk()
window.title("Drowsiness Detection System")

# Add Start and Stop buttons
start_button = tk.Button(window, text="Start Detection", command=start_detection, font=("Helvetica", 16), bg="green", fg="white")
start_button.pack(pady=20)

stop_button = tk.Button(window, text="Stop Detection", command=stop_detection, font=("Helvetica", 16), bg="red", fg="white")
stop_button.pack(pady=20)

# Add Exit button
exit_button = tk.Button(window, text="Exit", command=window.quit, font=("Helvetica", 16), bg="black", fg="white")
exit_button.pack(pady=20)

# Run the Tkinter event loop
window.mainloop()
```
